```python
import requests
import json
import pandas as pd
```

```python
import requests
from bs4 import BeautifulSoup

response = requests.get("https://tw.search.buy.yahoo.com/search/shopping/product?kw=%E9%9B%BB%E6%9A%96%E5%99%A8&p=%E9%9B%BB%E6%9A%96%E5%99%A8&cid=5&clv=1")
soup = BeautifulSoup(response.text, 'lxml')
```

```python
soup
```

### 使用正規式 搜尋文本中產品標籤文字內容

```python
print(soup.find_all("span", {"class":"BaseGridItem__title___2HWui"}))
```

```python
#soup.find("title").text

product_name = [tag.text for tag in soup.find_all("span", {"class":"BaseGridItem__title___2HWui"})]
product_name
```

### 使用正規式 搜尋文本中產品售價文字內容

```python
product_price = [tag.text for tag in soup.find_all("span", {"class":"BaseGridItem__price___31jkj"})]

product_price
```

## 將爬取到的資訊存成dataframe

```python
df_ = pd.DataFrame({"電暖器":product_name,
                             "價格":product_price,
                             }, columns = ["電暖器", "價格"])
df_ 
```

# 爬取氣象資料



```python
import pandas as pd
import numpy as np

# 氣象局提供之2017年每個月氣象之氣象資料
href = "https://www.cwb.gov.tw/V7/climate/monthlyData/Data/mD20172.htm"

#後面需要加上[0]  才取得到表格
df =  pd.read_html(href ,header = 1, thousands= ',',encoding ='lxml' )[0]
```

```python

prestr =  "https://www.cwb.gov.tw/V7/climate/monthlyData/Data/mD2017"
tailstr =  ".htm"


#讀取表格中 台北月均溫的columns
for r in np.arange(1,13,1):

    url = prestr + str(r) + tailstr
    df = pd.read_html(url)[0]
    print( str(r) + "月台北均溫 "+df.iloc[25,1])
```
