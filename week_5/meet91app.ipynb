# 做四分位距

```python
quantiles = rfmTable.quantile(q=[0.20,0.40,0.60,0.80])
quantiles
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
rfm = rfmTable
```

# x 是引入要比較的數值
# p 是遞迴值
# d 是先設好的quantiles字典

```python
def RClass(x,p,d):
    if x <= d[p][0.20]:
        return 1
    elif x <= d[p][0.40]:
        return 2
    elif x <= d[p][0.60]: 
        return 3
    elif x <= d[p][0.80]: 
        return 4
    else:
        return 5
    
def FMClass(x,p,d):
    if x <= d[p][0.20]:
        return 5
    elif x <= d[p][0.40]:
        return 4
    elif x <= d[p][0.60]:
        return 3
    elif x <= d[p][0.80]: 
        return 2
    else:
        return 1
```

# x 是引入要比較的數值
# p 是遞迴值
# d 是先設好的quantiles字典

```python
###### rfm.apply()
rfm['R_Quartile'] = rfm['Recency'].apply(RClass, args=('Recency',quantiles,))
rfm['F_Quartile'] = rfm['Frequency'].apply(FMClass, args=('Frequency',quantiles,))
rfm['M_Quartile'] = rfm['Monetary'].apply(FMClass, args=('Monetary',quantiles,))
```

```python
rfm['RFMClass'] = rfm.R_Quartile.map(str) \
                            + rfm.F_Quartile.map(str) \
                            + rfm.M_Quartile.map(str)
```

```python
def classifier(row):
    if row['RFMClass']=='555':
        return '重要價值客戶'
    elif row['RFMClass']=='155':
        return '重要喚回客戶'
    elif row['RFMClass']=='515':
        return '重要深耕客戶'
    elif row['RFMClass']=='115':
        return '重要挽留客戶'
    elif row['RFMClass']=='511':
        return '新客戶'
    elif row['RFMClass']=='111':
        return '流失客戶'
    else:
        return 'other'
```

```python
rfm['newclass'] = rfm.apply(classifier, axis=1)
```
