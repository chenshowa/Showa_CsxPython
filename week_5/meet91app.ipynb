```python
quantiles = rfmTable.quantile(q=[0.20,0.40,0.60,0.80])
quantiles
```

```python
quantiles = quantiles.to_dict()
quantiles
```

```python
rfm = rfmTable
```

# x ÊòØÂºïÂÖ•Ë¶ÅÊØîËºÉÁöÑÊï∏ÂÄº
# p ÊòØÈÅûËø¥ÂÄº
# d ÊòØÂÖàË®≠Â•ΩÁöÑquantilesÂ≠óÂÖ∏

```python
def RClass(x,p,d):
    if x <= d[p][0.20]:
        return 1
    elif x <= d[p][0.40]:
        return 2
    elif x <= d[p][0.60]: 
        return 3
    elif x <= d[p][0.80]: 
        return 4
    else:
        return 5
    
def FMClass(x,p,d):
    if x <= d[p][0.20]:
        return 5
    elif x <= d[p][0.40]:
        return 4
    elif x <= d[p][0.60]:
        return 3
    elif x <= d[p][0.80]: 
        return 2
    else:
        return 1
```

# x ÊòØÂºïÂÖ•Ë¶ÅÊØîËºÉÁöÑÊï∏ÂÄº
# p ÊòØÈÅûËø¥ÂÄº
# d ÊòØÂÖàË®≠Â•ΩÁöÑquantilesÂ≠óÂÖ∏

```python
###### rfm.apply()
rfm['R_Quartile'] = rfm['Recency'].apply(RClass, args=('Recency',quantiles,))
rfm['F_Quartile'] = rfm['Frequency'].apply(FMClass, args=('Frequency',quantiles,))
rfm['M_Quartile'] = rfm['Monetary'].apply(FMClass, args=('Monetary',quantiles,))
```

```python
rfm['RFMClass'] = rfm.R_Quartile.map(str) \
                            + rfm.F_Quartile.map(str) \
                            + rfm.M_Quartile.map(str)
```

```python
def classifier(row):
    if row['RFMClass']=='555':
        return 'ÈáçË¶ÅÂÉπÂÄºÂÆ¢Êà∂'
    elif row['RFMClass']=='155':
        return 'ÈáçË¶ÅÂñöÂõûÂÆ¢Êà∂'
    elif row['RFMClass']=='515':
        return 'ÈáçË¶ÅÊ∑±ËÄïÂÆ¢Êà∂'
    elif row['RFMClass']=='115':
        return 'ÈáçË¶ÅÊåΩÁïôÂÆ¢Êà∂'
    elif row['RFMClass']=='511':
        return 'Êñ∞ÂÆ¢Êà∂'
    elif row['RFMClass']=='111':
        return 'ÊµÅÂ§±ÂÆ¢Êà∂'
    else:
        return 'other'
```
