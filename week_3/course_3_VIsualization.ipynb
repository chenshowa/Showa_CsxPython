```python
%matplotlib inline
import requests
import json
import pandas as pd
import matplotlib.pyplot as plt


import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\chenhur\Documents\GitHub\Showa_CsxPython\Ntu_DataSet\Ntu_Orders.csv')
```

```python
df['new_DateId'] = pd.to_datetime(df['DateId'].astype(str), format='%Y%m%d')
```

```python
data_Quantity = df.groupby(by='new_DateId')['Quantity'].sum()
data_Quantity .plot() 
```

## 取貨地點統計 

```python
ShippingProfileType_data = df.groupby(by='ShippingProfileTypeDef')['Quantity'].sum()

ShippingProfileType_data.plot(kind='bar')

plt.show()
```

## 消費者端常用裝置

```python
ShippingProfileType_data = df.groupby(by='TrackSourceTypeDef')['Quantity'].sum()
ShippingProfileType_data.plot(kind='bar')
plt.show()
```
