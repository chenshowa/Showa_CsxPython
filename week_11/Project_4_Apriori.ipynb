### 分析商品和折扣類型之間的關聯性


```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import pandas as pd
```

```python
order = pd.read_csv(r'C:\Users\chenhur\Documents\GitHub\Showa_CsxPython\Ntu_DataSet\Ntu_Orders.csv')
promotion =  pd.read_csv(r'C:\Users\chenhur\Documents\GitHub\Showa_CsxPython\Ntu_DataSet\NTU_1317_Promotion.txt',sep='\t')  
promotionorders =pd.read_csv(r'C:\Users\chenhur\Documents\GitHub\Showa_CsxPython\Ntu_DataSet\NTU_1317_PromotionOrders.txt',sep='\t',low_memory=False)  

order = order.rename(columns=lambda x: x.replace('\ufeff', ''))
promotion = promotion.rename(columns=lambda x: x.replace('\ufeff', ''))
promotionorders = promotionorders.rename(columns=lambda x: x.replace('\ufeff', ''))
```

```python
PromotionConditions =pd.read_csv(r'C:\Users\chenhur\Documents\GitHub\Showa_CsxPython\Ntu_DataSet\PromotionConditions.csv')  
```

```python
promocondi = pd.merge(promotionorders, PromotionConditions, on = 'PromotionId', how = 'left')
ordpromocondi= pd.merge(order, promocondi, on = 'SalesOrderSlaveId', how = 'left')
```

```python
df_two = ordpromocondi[['TradesOrderGroupCode','LevelFourCategoryName','DiscountType']]  
df_two['quantity'] = 1
df_two
```

```python
df_test = df_two
df_test['merge'] = df_two['LevelFourCategoryName'] + df_two['DiscountType']
df_test
```

```python
LevelFour = df_two[['TradesOrderGroupCode','LevelFourCategoryName']] 
LevelFour.columns = ['TradesOrderGroupCode','DiscountType']

DiscountType = df_two[['TradesOrderGroupCode','DiscountType']] 
```

```python
FourDiscountType = pd.concat([LevelFour, DiscountType], axis=0)
FourDiscountType.dropna()
```

```python
FourDiscountType['quantity'] = 1
FourDiscountType = FourDiscountType.groupby(by=['TradesOrderGroupCode','DiscountType'])['quantity'].sum()
FourDiscountType
```

```python
FourDiscountType = FourDiscountType.unstack().fillna(0).applymap(lambda x: 1 if x>0 else 0)
FourDiscountType = FourDiscountType.drop('T恤/帽T',axis=1)
FourDiscountType
```

Support : Between 0.1 and 0.2

Confidence : Between 0.5 and 0.7

```python

FourDiscountType_frequency = apriori(FourDiscountType, min_support=0.01, use_colnames=True)     
FourDiscountType_rules = association_rules(FourDiscountType_frequency, metric="confidence", min_threshold=0.4)
FourDiscountType_rules.sort_values(by = 'confidence',ascending=False)
```

### 不管先出現的項目是甚麼，' 滿額折現 ' 總是會出現在後頭，' 滿額折現 ' 似乎是店家最常用的折扣類型，

- confidence(A→C) = support(A→C) / support(A)

- lift(A→C)      = confidence(A→C)/support(C)----------------support(C)發生的情況下confidence(A→C)

- levarage(A→C)   = support(A→C) − support(A)×support(C)------------A C 獨立發生 與AC關聯發生的機率

- conviction(A→C) = 1 − support(C) / 1 − confidence(A→C)--------------數值高意味 後者很大程度上取決於先行者，如果是獨立的，則為1

```python
# 稍微調個參數
FourDiscountType_frequency = apriori(FourDiscountType, min_support=0.016, use_colnames=True)     
FourDiscountType_rules = association_rules(FourDiscountType_frequency, metric="confidence", min_threshold=0.6)
FourDiscountType_rules.sort_values(by = 'confidence',ascending=False)
```

### '雪紡/棉麻衫' 常和 '滿額折現' 一起出現 

```python
FourDiscountType = FourDiscountType.drop('滿額折現',axis=1)
```

### ' 滿額折現 ' 是店家最常用的折扣類型，因此予以刪除

```python
FourDiscountType_frequency = apriori(FourDiscountType, min_support=0.012, use_colnames=True)
FourDiscountType_rules = association_rules(FourDiscountType_frequency, metric="confidence", min_threshold=0.4)
FourDiscountType_rules.sort_values(by = 'confidence',ascending=False)
```

### 第二項(滿件打折, 滿額贈) 與(針織衫)的關聯度，雖然兩者的confidence小於前一項，但lift和conviction 為最高

```python
FourDiscountType = FourDiscountType.drop('背心/小可愛',axis=1)
```
