- confidence(A→C) = support(A→C) / support(A)

- lift(A→C)      = confidence(A→C)/support(C)----------------support(C)發生的情況下confidence(A→C)

- levarage(A→C)   = support(A→C) − support(A)×support(C)------------A C 獨立發生 與AC關聯發生的機率

- conviction(A→C) = 1 − support(C) / 1 − confidence(A→C)--------------數值高意味 後者很大程度上取決於先行者，如果是獨立的，則為1

### 同一購物車的商品放在一起, 且同一購物車同一品項數量相加總

```python
df_new = order[['TradesOrderGroupCode','LevelFourCategoryName']]  
df_new['quantity'] = 1
df_new
```

```python
df_new_sum = df_new.groupby(by=['TradesOrderGroupCode','LevelFourCategoryName'])['quantity'].sum()
df_new_sum
```

### 維度1為購物車   ------   維度2為品項   

### unstack  把資料變map       fillna  把nan的空格填滿 0       .applymap 把空格內大於0的值都變成1

```python
df_new_one_hot = df_new_sum.unstack().fillna(0).applymap(lambda x: 1 if x>0 else 0)
df_new_one_hot
```

```python
# T恤/帽T 是普遍大家都會買的東西

df_new_one_hot = df_new_one_hot.drop('T恤/帽T',axis=1)
df_new_one_hot
```

```python
frequent_itemsets = apriori(df_new_one_hot, min_support=0.01, use_colnames=True)     
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.35)

rules.sort_values(by = 'confidence',ascending=False)
```
